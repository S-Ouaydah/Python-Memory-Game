```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from PIL import ImageTk,Image
from time import sleep
import pprint

class Card(Frame):
    new_width=0
    new_height=0
    
    def __init__(self, master,img, gridrow=0, gridcol=0, *pargs):
        Frame.__init__(self, master,borderwidth=5,relief="ridge", *pargs)
        self.flipped = False
        self.gridrow = gridrow
        self.gridcol = gridcol
        self.master = master
        self.back = Image.open('images/card_back.png')
        self.front = img
        self.image = self.back
        self.photo = ImageTk.PhotoImage(self.image,master = master)

        self.button = Button(self, image=self.photo, command=self.flip_image, width=200, height=400)
        self.button.pack(fill=BOTH, expand=YES)
        self.button.bind('<Configure>', self.resize_image)
#         self.label.bind('<ButtonPress-1>', self.flip_image)

    def resize_image(self,event):
        global new_width,new_height
        new_width = event.width
        new_height = event.height
        
        self.photo = ImageTk.PhotoImage(self.image.resize((new_width, new_height)))
        self.button.configure(image = self.photo)
        
    def flip_image(self):
        if not self.flipped:
            if num_flipped < 1:
                self.flip()
            elif num_flipped < 2:
                self.flip()
                self.check_flipped()

            else:
                pass
        else:
            pass

        
    def flip(self):
        global num_flipped
        self.flipped = not self.flipped
        cards_flipped.append(self)
        num_flipped += 1
        self.image = self.front
        
        self.photo = ImageTk.PhotoImage(self.image.resize((new_width, new_height)))
        self.button.configure(image = self.photo)
        self.update()
        
    def flip_back(self):
        global num_flipped
        self.flipped = not self.flipped
        num_flipped -= 1
        self.image = self.back
        
        self.photo = ImageTk.PhotoImage(self.image.resize((new_width, new_height)))
        self.button.configure(image = self.photo)
    def check_flipped(self):
        global num_flipped,score,target_score
        if len(cards_flipped) > 2:
            print("You shall not flip more than 2 cards!!!")
            return
        if cards_flipped[0].front == cards_flipped[1].front:
            cards_flipped[0].button.configure(state='disabled')
            cards_flipped[1].button.configure(state='disabled')
            num_flipped=0
            score += 10
            if score == target_score:
                messagebox.showinfo("Congratz!","You completed this difficulty!")
                
        else:
            sleep(1)
            for card in cards_flipped:
                card.flip_back()
                card.update()
        del cards_flipped[:]
        
#             self.after(8000,self.flip_back())
#             Card.ff_card.after(8000,Card.ff_card.flip_back())
```

```python
from random import shuffle

card_imgs = [Image.open('images/card'+str(i+1)+'.png') for i in range(22)]
current_dif = 'Basic'

def change_dif(event):
    global current_dif,score
    diff = difficulty.get()
    if diff != current_dif:
        current_dif = diff
        for item in frame.winfo_children():
            item.destroy()
        generate_cards(diff)
        score=0
        
def reconfigure(tup):
    rows,cols = tup
    for row in range(rows):
        frame.rowconfigure(row, weight=1)
    for col in range(cols):
        frame.columnconfigure(col, weight=1)
    
def generate_cards(dif):
    global target_score
    shuffle(card_imgs)
    if dif=='Basic':        
        reconfigure(gen_lvl(12,6))
        target_score = 60
    if dif=='Simple':        
        reconfigure(gen_lvl(24,12))
        target_score = 120
    if dif=='Easy':        
        reconfigure(gen_lvl(40,10))
        target_score = 200
        
def gen_lvl(tot,cols):
    selected=[card_imgs[i] for i in range(tot//2) for __ in range(2)]
    shuffle(selected)
    card_widgets = [Card(frame, selected[i]) for i in range(tot)]
    col=0
    row=0
    for card in card_widgets:
        card.grid(row=row,column=col)
        col+=1
        if(col==cols):
            col=0
            row+=1
    return (row,cols)
    
```
